```python
from dotenv import load_dotenv
load_dotenv()

from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="deepseek-v4-pro", base_url="https://api.deepseek.com")

messages = [
    SystemMessage(content="你是一个中英翻译助手，帮助用户将中文翻译成英文，或者将英文翻译成中文。"),
    HumanMessage(content="LangChain provides a prebuilt agent architecture and model integrations to help you get started quickly and seamlessly incorporate LLMs into your agents and applications."),
]

model.invoke(messages)
```
